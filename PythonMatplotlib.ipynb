```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Basic Graph

matplotlib.pyplot

matplotlib.pyplot â€” Matplotlib 3.5.3 documentation

plot(*args[, scalex, scaley, data])

```python
# Make Basic Graph.
x = [0,1,2,3,4,5]#X-Axis-Values.
y = [0,2,4,6,8,10]#Y-Axis-Values.
plt.plot(x,y,
         label='2x',
         color='red',
         linewidth=2,
         marker='.',
        markersize=10,
         markeredgecolor='blue',
        linestyle='--')#(X,Y).
#>>Graph-Title<<.
plt.title('Our First Graph',
          fontdict={'fontname':'Comic Sans MS',
         'fontsize':20,'color':'red'})
#>>Label-Text<<.
plt.xlabel('X Axis',fontdict={'fontname':'Comic Sans MS'})
plt.ylabel('Y Axis',fontdict={'fontname':'Comic Sans MS'})
#-------------------------
plt.xticks([0,1,2,3,4,5])
plt.yticks([0,2,4,6,8,10])
#-------------------------
plt.legend()
plt.show()
```

```python
x = [0,1,2,3,4,5]
y = [0,2,4,6,8,10]
# Resize Your Graph.
plt.figure(figsize=(5,3),dpi=300)#Figure-Size.
plt.plot(x,y,'b^--',label='2xAhmed')
## Line Number Two.
x2 = np.arange(0,4.5,0.5)
print(x2)
#>>Draw Line on Graph.
plt.plot(x2[:6],x2[:6]**2,'r',label='X^2')#x2,x2Power2.
plt.plot(x2[5:],x2[5:]**2,'r--')
#---------------------------
plt.title('Our Fisrt Graph',
fontdict={'fontname' : 'Arial',
'fontsize':20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
#-------------------
plt.xticks([0,1,2,3,4,5])
plt.yticks([0,2,4,6,8,10])
#-------------------
plt.legend()#2xAhmed
plt.savefig('mygraph.png',dpi=300)
plt.show()
```

```python
labels = ['A','B','C']
values = [1,4,2]
bars=plt.bar(labels,values)#(X-Axis,Y-Axis).
#bars[0].set_hatch('/')
#bars[1].set_hatch('O')
#bars[2].set_hatch('*')
patterns = ['/','O','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
plt.figure(figure=(6,4))
plt.show()
```
